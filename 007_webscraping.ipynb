```python
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen

link = "https://scriptureearth.org/data/bom/sab/bom/bom-22-SNG-001.html"
req = Request(link, headers={'User-Agent': 'Mozilla/5.0'})
url = urlopen(req).read()
soup = BeautifulSoup(url, "lxml")

# Find the parent div
parent_div = soup.find('div', class_='sp') 

# Remove the parent div and its child subdivision div with id 'sp1'
if parent_div:
    div_ids_to_remove = ['sp1', 'sp2', 'sp3', 'sp4', 'sp5']
    for div_id in div_ids_to_remove:
        child_div = soup.find('div', id=div_id)
        if child_div:
            child_div.extract()
        
# Print the remaining content
text = soup.text

print(text)
```
